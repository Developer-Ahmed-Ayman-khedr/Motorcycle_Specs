```python
import pandas as pd
import numpy as np
```

```python
filepath = "SourceData/all_bikez_curated.csv"
df = pd.read_csv(filepath)
df.head(10)
```

```python
df.describe()
```

```python
print(df.dtypes)
```

```python
print(df.info())
```

```python
print(df.isna().sum())
```

```python
df.replace('?',np.nan,inplace=True)
df.replace('0',np.nan,inplace=True)
```

```python
# drop the row with missing model
df.dropna(subset=['Model'], axis=0, inplace=True)
print(df.isna().sum())
```

```python
df.head(5)
```

```python
# drop the row with missing Engine stroke
df.dropna(subset=['Engine stroke'], axis=0, inplace=True)
print(df.isna().sum())
```

```python
df.head(15)
```

```python
# Work on Color options column
print(df['Color options'].isna().sum())
print(df['Color options'].unique())
```

```python
# Get the most common color
common_color = df['Color options'].value_counts().idxmax()

# Replace the NaN values with the common color value
df['Color options'].replace(np.nan, common_color, inplace=True)

# Check if the changes worked
print(df['Color options'].isna().sum())
```

```python
# Find the columns with Nan values
print(df.isna().sum())
```

```python
# Work on rear suspension column
print(df['Rear suspension'].isna().sum())
print(df['Rear suspension'].unique())

# Get the most common rear suspension
common_rear_suspension = df['Rear suspension'].value_counts().idxmax()
print(common_rear_suspension)

# Replace the NaN values with the common rear suspension value
df['Rear suspension'].replace(np.nan, common_rear_suspension, inplace=True)

# Check if the changes worked
print(df['Rear suspension'].isna().sum())
```

```python
# Find the columns with Nan values
print(df.isna().sum())
df.head(5)
```

```python
# Find the columns with Nan values
print(df.isna().sum())
df.head(5)
```
