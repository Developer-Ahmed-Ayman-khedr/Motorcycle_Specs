```python
def Wordk_on_columns(column_name, Data_Type):
    # Show the column before changing
    print("-----------------------------------------------------------------------------------------------------------------\n")
    print("Before changes:\n")
    print("Number of NaN values",df[column_name].isna().sum())
    print("-------------------------------------------------\n")
    print("Unique values",df[column_name].unique())
    #object Data types
    if Data_Type=='type_object':
        # Get the most common Cooling system
        common_value = df[column_name].value_counts().idxmax()
        print("-------------------------------------------------\n")
        print("Common value: ",common_value)
        
        # Replace the NaN values with the commo value
        df[column_name].replace(np.nan, common_value, inplace=True)
    #float & int Data types
    elif (Data_Type=='type_int64') or (Data_Type=='type_float64'):
        # Get the average value
        avg_Wheelbase = df[column_name].astype('float').mean(axis=0)
        print("-------------------------------------------------\n")
        print("average value: ",avg_Wheelbase.astype('int64'))
        
        # Replace the NaN values with the average value
        df[column_name].replace(np.nan, avg_Wheelbase.astype('int64'), inplace=True)
        
    # Check if the changes worked
    print("-----------------------------------------------------------------------------------------------------------------\n")
    print("After changes:\n")
    print("Number of NaN values: ",df[column_name].isna().sum())
    print("Unique values",df[column_name].unique())
        
    # Find the columns with Nan values
    print("-----------------------------------------------------------------------------------------------------------------\n")
    print("Show the columns:\n")
    print(df.isna().sum())
    print("-------------------------------------------------\n")
    df.head(5)
```
