```python
df.replace('?',np.nan,inplace=True)
df.replace('0',np.nan,inplace=True)
```

```python
# drop the row with missing model
df.dropna(subset=['Model'], axis=0, inplace=True)
print(df.isna().sum())
```

```python
# drop the row with missing Engine stroke
df.dropna(subset=['Engine stroke'], axis=0, inplace=True)
print(df.isna().sum())
```

```python
df.head(5)
```

```python
def Wordk_on_columns(column_name, Data_Type):
    # Show the column before changing
    print("-----------------------------------------------------------------------------------------------------------------\n")
    print("Before changes:\n")
    print("Number of NaN values",df[column_name].isna().sum())
    print("-------------------------------------------------\n")
    print("Unique values",df[column_name].unique())
    #object Data types
    if Data_Type=='type_object':
        # Get the most common Cooling system
        common_value = df[column_name].value_counts().idxmax()
        print("-------------------------------------------------\n")
        print("Common value: ",common_value)
        
        # Replace the NaN values with the commo value
        df[column_name].replace(np.nan, common_value, inplace=True)
    #float & int Data types
    elif (Data_Type=='type_int')&(Data_Type=='type_float'):
        # Get the average value
        avg_Wheelbase = df['Wheelbase (mm)'].astype('float').mean(axis=0)
        print("-------------------------------------------------\n")
        print("average value: ",avg_Wheelbase.astype('int64'))
        
        # Replace the NaN values with the average value
        df['Wheelbase (mm)'].replace(np.nan, avg_Wheelbase.astype('int64'), inplace=True)
        
    # Check if the changes worked
    print("-----------------------------------------------------------------------------------------------------------------\n")
    print("After changes:\n")
    print("Number of NaN values: ",df[column_name].isna().sum())
    print("Unique values",df['Cooling system'].unique())
        
    # Find the columns with Nan values
    print("-----------------------------------------------------------------------------------------------------------------\n")
    print("Show the columns:\n")
    print(df.isna().sum())
    print("-------------------------------------------------\n")
    df.head(5)
```

```python
Wordk_on_columns('Color options', 'type_object')
```

```python
Wordk_on_columns('Rear suspension', 'type_object')
```

```python
Wordk_on_columns('Front suspension', 'type_object')
```
